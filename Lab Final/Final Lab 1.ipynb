```python

# Question 1

from PIL import Image
 
size = (200, 200)
img = Image.open(r"new1.jpg")
 
    
# Question 2
print("Original size of the image")
print(img.size)

r_img = img.resize(size, resample = Image.BILINEAR)
 
    
    
# Question 3
r_img.save("resized.jpeg")
 
img = Image.open(r"resized.jpeg")

print(img.size)

```

```python
import cv2
from skimage import io, img_as_float
from skimage.restoration import denoise_tv_chambolle
from matplotlib import pyplot as plt

img  = img_as_float(io.imread("new1.jpg", as_gray=True))
plt.hist(img.flat, bins=100, range=(0,1))

denoise_img = denoise_tv_chambolle(img, weight=0.1, eps=0.0002, channel_axis=False)

plt.hist(denoise_img.flat, bins=100, range=(0,1))

cv2.imshow("Original", img)
cv2.imshow("TV Filtered", denoise_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
