### Task 1

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('ramadan.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('pasta.jpg', cv2.IMREAD_GRAYSCALE)

canny1 = cv2.Canny(img1, 100, 200)
canny2 = cv2.Canny(img2, 100, 200)

sobelx1 = cv2.Sobel(img1, cv2.CV_64F, 1, 0, ksize=3)
sobely1 = cv2.Sobel(img1, cv2.CV_64F, 0, 1, ksize=3)
sobel1 = np.sqrt(sobelx1**2 + sobely1**2)
sobelx2 = cv2.Sobel(img2, cv2.CV_64F, 1, 0, ksize=3)
sobely2 = cv2.Sobel(img2, cv2.CV_64F, 0, 1, ksize=3)
sobel2 = np.sqrt(sobelx2**2 + sobely2**2)

plt.figure(figsize=(12, 8))
plt.subplot(2, 4, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original image 1')
plt.subplot(2, 4, 2)
plt.imshow(canny1, cmap='gray')
plt.title('Canny edge image 1')

plt.subplot(2, 4, 3)
plt.imshow(sobel1, cmap='gray')
plt.title('Sobel edge image 1')
plt.subplot(2, 4, 4)
plt.imshow(img1 - canny1 - sobel1, cmap='gray')
plt.title('Difference image 1')
plt.subplot(2, 4, 5)
plt.imshow(img2, cmap='gray')
plt.title('Original image 2')
plt.subplot(2, 4, 6)
plt.imshow(canny2, cmap='gray')
plt.title('Canny edge image 2')
plt.subplot(2, 4, 7)
plt.imshow(sobel2, cmap='gray')
plt.title('Sobel edge image 2')
plt.subplot(2, 4, 8)
plt.imshow(img2 - canny2 - sobel2, cmap='gray')
plt.title('Difference image 2')
plt.show()
```

### Task 2

```python
import cv2
import matplotlib.pyplot as plt
from scipy.fft import fft2, fftshift

img1 = cv2.imread('ramadan.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('pasta.jpg', cv2.IMREAD_GRAYSCALE)

fft1 = fft2(img1)
fft2 = fft2(img2)

f_shift1 = fftshift(fft1)
f_shift2 = fftshift(fft2)

mag1 = np.abs(f_shift1)
mag2 = np.abs(f_shift2)

plt.figure(figsize=(12, 6))

plt.subplot(221), plt.imshow(img1, cmap="gray"), plt.title("Image 1 (Original)")
plt.subplot(222), plt.imshow(mag1, cmap="gray"), plt.title("Magnitude of FT (Image 1)")

plt.subplot(223), plt.imshow(img2, cmap="gray"), plt.title("Image 2 (Original)")
plt.subplot(224), plt.imshow(mag2, cmap="gray"), plt.title("Magnitude of FT (Image 2)")

plt.tight_layout()
plt.show()
```
