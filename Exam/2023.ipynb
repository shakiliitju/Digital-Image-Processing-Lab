```python
from matplotlib import pyplot as plt
import cv2

img1  = cv2.imread("parrot-webp.webp")
img2  = cv2.imread("gray.jpeg")

img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

merged = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)

plt.imshow(merged)
```

```python

from skimage import io
from matplotlib import pyplot as plt
from skimage.filters import gaussian, sobel
import cv2

img = io.imread("gray.jpeg")
plt.imshow(img)

gaussian_using_skimage = gaussian(img, sigma=1, mode='constant', cval=0.0)
plt.imshow(gaussian_using_skimage)

img_gray = io.imread("gray.jpeg")
sobel_img = sobel(img_gray)
plt.imshow(sobel_img, cmap='gray')
```
