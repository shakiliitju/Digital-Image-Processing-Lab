## RGB to CMY Convention

```python
import cv2
print("Original Image")
image = cv2.imread("/content/Shakiliitju.jpg")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print("RGB Image")
R,G,B = cv2.split(image)
print("R")
cv2_imshow(R)

print("G")
cv2_imshow(G)

print("B")
cv2_imshow(B)


C=1-R
M=1-G
Y=1-B
print("CMY image")
print("C Complacent")
cv2_imshow(C)

print("C Complacent")
cv2_imshow(M)

print("C Complacent")
cv2_imshow(Y)
```
