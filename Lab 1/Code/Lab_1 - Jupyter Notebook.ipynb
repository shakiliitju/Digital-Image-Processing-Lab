## RGB to CMY Convention

```python
import cv2
print("Original Image")
image = cv2.imread("shakiliitju.jpg")
image
print("RGB Image")
R,G,B = cv2.split(image)
print("R")
R

print("G")
G

print("B")
B


C=1-R
M=1-G
Y=1-B
print("CMY image")
print("C Complacent")
C

print("C Complacent")
M

print("C Complacent")
Y
```
